```python
#Furnishing Status and their count

housing_['furnishingstatus'].value_counts()
```

```python
#Averge Price of House in Main Roads.

avg_price = housing_.loc[housing_['mainroad']=='yes', 'price'].mean()
avg_price_int = int(avg_price)
print(avg_price_int)

```

```python
#Houses whose area >10000.

housing_.loc[(housing_['area']>10000)]
```

```python
#Furnished Houses whose area > 10000.

housing_.loc[(housing_['area']>10000)&(housing_['furnishingstatus']=='furnished')]
```

```python
#aggregation of Furnishing status of houses

result = housing_.groupby(['furnishingstatus']).agg({'price':['mean','max'],'area':'mean'}).reset_index()

result.columns = ['furnishing_status', 'price_mean', 'pricemax', 'area_mean']

result['price_mean'] = result['price_mean'].round(0).astype(int)
result['price_max'] = result['pricemax'].astype(int)
result['area_mean'] = result['area_mean'].round(0).astype(int)

print (result)
```

```python

#Calculate the mean price for each furnishing status group whose area >5000.


result=housing_.loc[(housing_['area']>5000),:].groupby(['furnishingstatus']).agg({'price':'mean'}).reset_index()
result.columns=['furnishung','price_mean']
result['price_mean']=result['price_mean'].round(0).astype(int)
print(result)
```

```python

#Calculate the mean price for each furnishing status group.

result1=housing_.groupby(["furnishingstatus"]).agg({'price':'mean'}).reset_index()
result1['price']=result1['price'].astype(int)

print(result1)
```

```python
#Average price of houses where number of rooms is greater than 3.


housing_[housing_["bedrooms"]>3]["price"].mean().astype(int)
```

```python
# Total number of houses for each furnishingstatus & stories


housing_.groupby(['furnishingstatus', 'stories']).size().reset_index(name="count")


```

```python

#List the top 5 houses with highest area but only for 'semi-furnished' houses.

avg=housing_.loc[housing_['furnishingstatus']=='semi-furnished'].sort_values(by='area',ascending=False).head(5)
print(avg.reset_index())

```

```python
top5_semi = housing_.loc[housing_['furnishingstatus']=='semi-furnished'].sort_values(by='area', ascending=False).head(5)
print(top5_semi)
```

```python
#Find the minimum, maximum, and average area of houses having airconditioning but no guestroom.

filtered=housing_.loc[(housing_['airconditioning']=='yes')&(housing_['guestroom']=='no'),'area'].mean().max().min()
print('max :',filtered.max())
print('mean:',filtered.mean())

print('min :',filtered.min())

```
